```python
import random


print("\n===> Number Guessing Game <===")
print("\nI have a number in my mind between 1-100.")
print("\nAttempts: 10")


secret = random.randint(1,100)
score = 0
attempts = 10
count = 0


while count < attempts:
    try:
        guess = int(input("\nEnter a number between 1-100...").strip())
        diff = abs(secret - guess)
        if diff == 0:
            print("\n😎 Perfect! Your Guess is right.")
            print(f"\nYou took {count + 1} attempts to guess the number.")
            print(f"\nYour score is {1000 - ((count+1)*100)}.")
            break
        elif diff <= 3:
            print("\nToo Close!")
        elif diff <= 7:
            print("\nHot!")
        elif diff <= 15:
            print("\nCool!")
        else:
            print("\nCold!")
    except ValueError:
        print('\nInvaild number. Enter number 1-50!')

    count+=1

if count == attempts and guess != secret:
    print(f"\nYou are out of attempts. The secret was {secret}")
    print(f"\nYour score is {score}.")
```
